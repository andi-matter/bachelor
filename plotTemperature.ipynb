```python
import numpy as np
import matplotlib.pyplot as plt


import math
import csv
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

import matplotlib as mpl

#matplotlib.verbose.level = 'debug-annoying'
mpl.rcParams.update({'font.size': 16})
import sys
import os
```

```python
temperatureFile = "/mnt/d/RUNDATA_UNI/temperature/Temperature Detector Lab-data-as-seriestocolumns-2021-04-23 10 24 45.csv" #sys.argv[1]
#saveFolder = sys.argv[2]

df = pd.read_csv(temperatureFile, skiprows=1)
df["Mean w/o T3"] = 1/5*(df["T0"] + df["T1"] + df["T2"] + df["T4"] + df["T5"])
keys = df.columns
print(keys)
df.head()
```

```python
fig = go.Figure()
for key in keys[1:-1]:
    fig.add_trace(go.Scatter(x=df["Time"], y=df[key], mode='lines', name=key))
fig.add_trace(go.Scatter(x=df["Time"], y=df["Mean w/o T3"], name="Mean w/o T3", line=dict(color='black', width=4, dash='dot')))
fig.update_layout(title="Temperature over time", xaxis_title="Date", yaxis_title="Temperature (deg.)")
fig.show()
if not os.path.exists("temperatures"):
    os.mkdir("temperatures")
fig.write_image("temperatures/temperatures.pdf")

```
