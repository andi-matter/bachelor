```python
import numpy as np
import matplotlib.pyplot as plt


import math
import csv
import pandas as pd
import plotly.express as px

import plotly.graph_objects as go

import matplotlib as mpl

#matplotlib.verbose.level = 'debug-annoying'
mpl.rcParams.update({'font.size': 16})
import sys
import os
```

```python
temperatureFile = "/mnt/d/RUNDATA_UNI/temperature/Temperature Detector Lab-data-as-seriestocolumns-2021-08-31 16 52 05.csv" #sys.argv[1]
#saveFolder = sys.argv[2]

df = pd.read_csv(temperatureFile, skiprows=1)
keys = df.columns
float_keys = []
for key in keys[1:]:
    df[key+"_float"] = np.zeros(len(df))
    float_keys.append(key+"_float")

for i in range(len(df)):
    for key in keys[1:]:
        end = str(df[key][i]).find('°') - 1
        #print(df[key][i])
        try:
            df[key+"_float"][i] = float(str(df[key][i])[:end])
        except ValueError:
            #print("Missing values")
            df[key+"_float"][i] = None
            continue

df["Mean"] = 1.0/3*(df["T0_float"] + df["T1_float"] + df["T2_float"]) # + df["T4"] + df["T5"])
keys = df.columns
#print(keys)
df["T0"][0]
```

```python
fig = go.Figure()
for key in float_keys:
    fig.add_trace(go.Scatter(x=df["Time"], y=df[key], mode='markers', name=key))
fig.add_trace(go.Scatter(x=df["Time"], y=df["Mean"], name="Mean", mode='markers', marker_color='rgba(0,0,0,1)'))
fig.update_traces(marker_size=4)
fig.update_layout(title="Temperature over time", xaxis_title="Date", yaxis_title="Temperature (°C)")
fig.show()
#if not os.path.exists("temperatures"):
   # os.mkdir("temperatures")
#fig.write_image("temperatures/temperatures.pdf")

```
